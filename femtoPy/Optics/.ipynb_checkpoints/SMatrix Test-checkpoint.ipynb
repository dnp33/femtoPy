# 3 layer transmission test

```python
def sell3(lmda,coef):
    lmda2=lmda*lmda
    n=coef[0]+coef[1]/(lmda2-coef[2])-coef[3]*lmda2-coef[4]*lmda2*lmda2
    return np.sqrt(n)

# load/trim frequency dependent index for Al
dat=np.loadtxt('Al.txt')
lmda=dat[:,0]
MIN=np.amin(np.where(lmda > 0.3)); MAX=np.amin(np.where(lmda > 1.69))

# calc wavevector
k0=2*np.pi/lmda[MIN:MAX]
nAl=dat[MIN:MAX,1]+1j*dat[MIN:MAX,2]
lmda=lmda[MIN:MAX]

# calculate MgO index
MgO=[2.956362,0.02195770,0.01428322,0.01062387,0.0000204968]
nMgO=sell3(lmda,MgO)

# find dielectric functions
epsMgO=nMgO**2
epsAl=nAl**2

# layer thicknesses
l=[0.02,0.2,0.02]

#arrays to hold R/T
t=np.empty(k0.size,dtype=complex)
r=np.empty(k0.size,dtype=complex)

# loop over wavevectors
for i in range(k0.size):
    eps_r=[epsAl[i],epsMgO[i],epsAl[i]]
    film=ml.multiLayer(eps_r=eps_r,l=l,k0=k0[i])
    t[i]=film.S12[1,1]
    r[i]=film.S11[1,1]

# calculate power trans. & refl. coefs.
T=t*np.conjugate(t)
R=r*np.conjugate(r)

# plot
fig,ax=plt.subplots(figsize=(10,7))
ax.plot(lmda,T)
ax.plot(lmda,R)

ax.set_xlabel(r'wavevector ($\mu m^{-1}$)')
ax.set_ylabel('T,R')
```

# Single layer test with comparison to multiple beam interference calculation

```python
eps_r=[4]; k0=1; kx=0; ky=0

l=np.linspace(0,10,1000)
T_S=np.empty(l.size,dtype=complex)
for i in range(l.size):
    film=multiLayer(eps_r=[4],l=[l[i]],k0=k0)
    T_S[i]=film.S12[1,1]

def singleLayer(n,l,k):
    r=(n-1)/(1+n)
    t=2/(1+n)
    return t**2*n*np.exp(1j*n*l*k)/(1-r*r*np.exp(2*1j*n*l*k))

T_MB=singleLayer(2,l,1)

fig,ax=figure()
ax.plot(l,T_S*np.conjugate(T_S),label='S-Matrix')
ax.plot(l,T_MB*np.conjugate(T_MB),linestyle='--',label='thin-film formula')

ax.legend(loc=1)

ax.set_ylabel('Energy Transmission')
ax.set_xlabel('thickness (a.u)')

fig,ax=figure()
ax.plot(l,T_S-T_MB)

ax.set_xlabel('thickness (a.u)')
ax.set_ylabel('difference')
```

# non-dissipative multilayer transmission

```python
n1=4; l1=n1
n2=9; l2=n2/4
eps_r=[n1,n2,n1,n2]; l=[l1,l2,l1,l2]; kx=0; ky=0
V_air=-1j*np.matrix([[kx*ky,1+ky**2],[-kx*kx-1,-kx*ky]],dtype=complex)

k0=np.linspace(1.1,2,100)

T=np.empty(k0.size,dtype=complex)
R=np.empty(k0.size,dtype=complex)
for i in range(k0.size):
    film=multiLayer(eps_r=eps_r,l=l,k0=k0[i])
    T[i]=film.S12[1,1]
    R[i]=film.S11[1,1]
    
T=T*np.conjugate(T)
R=R*np.conjugate(R)
    
fig,ax=figure()
ax.plot(k0,T)
ax.plot(k0,R)

ax.plot(k0,T+R,color='k')

ax.set_xlabel('wavevector (a.u)')
ax.set_ylabel('T,R,T+R')

fig,ax=figure()
ax.plot(k0,1-(T+R),color='k')
ax.set_xlabel('wavevector (a.u)')
ax.set_ylabel('1-(T+R)')
```
